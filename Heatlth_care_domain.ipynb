```python
def classify_fetature(df):
  categorical_features = []
  non_categorical_features = []
  discrete_features = []
  continuous_features = []

  for i in df.columns:
    if df[i].dtype == 'object':
      if df[i].nunique()<10:
         categorical_features.append(i)
      else:
          non_categorical_features.append(i)

    elif df[i].dtype in ['int','float']:
      if df[i].nunique()<10:
        discrete_features.append(i)
      else:
        continuous_features.append(i)

  return categorical_features, non_categorical_features, discrete_features, continuous_features

# this is called binning / bundle
```

```python
categorical_features, non_categorical_features, discrete_features, continuous_features = classify_fetature(df)
```
